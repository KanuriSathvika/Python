```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.array([10,20,40])
y=np.array([30,20,50])
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y,"o")
```

```python
plt.plot(x,y,'black')
```

```python
plt.plot(x,y,'k')
```

```python
plt.plot(x,y,'p:r',markersize=12)

# A format string consists of a part for color, marker and line::
#     fmt = '[marker][line][color]'
```

### create an array with 10 points , ploting corresponding points such that the line is dotted , color of the marker is pink ,and marker fsce color will be purple

```python
arr=np.array([12,34,11,30,56,23,76,13,76,90])
plt.plot(arr,'.:',color='pink',mfc='purple',ms=20,linewidth=3)
#mfc-markerfacecolor:color of marker
#ms-markersize:size of marker
```
