```python
import pandas as pd
import numpy as np
```

```python
d1={'USN':[101,102,103,104],'Name':['tom','jerry','mickey','dora']}
df1=pd.DataFrame(d1)
df1
```

```python
d2={'USN':[101,103],'marks':[87,90]}
df2=pd.DataFrame(d2)
df2
```

```python
#Inner merge
pd.merge(df1,df2,how='inner',on='USN')
#It joins the common USN rows
```

```python
pd.merge(df1,df2,how='left',on='USN')
```

```python
pd.merge(df1,df2,how='right',on='USN')
```

```python
pd.merge(df1,df2,how='outer',on='USN')
```

```python
df1.join(df2)
```
